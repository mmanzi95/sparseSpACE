```python
# dimension
dim = 5

# domain
a = np.zeros(dim)
b = np.ones(dim)

# grid
grid=TrapezoidalGrid(a, b, dim)

# function
midpoints = np.ones(dim) * 0.99
coefficients = np.array([ 10**2 * (d+1) for d in range(dim)])
f = Function2Jakeman(midpoints,coefficients)

# reference solution for the integration
reference_solution = f.getAnalyticSolutionIntegral(a,b)

# error estimator
errorOperator=ErrorCalculatorExtendSplit()

# operation=integration
operation = Integration(f=f, grid=grid, dim=dim, reference_solution=reference_solution)

# initialize Extend-Split Strategy
adaptiveCombiInstanceExtend2 = SpatiallyAdaptiveExtendScheme(a, b, 5, grid, no_initial_splitting=False, version=0, automatic_extend_split=True, split_single_dim=False, operation=operation)
adaptiveCombiInstanceExtend = SpatiallyAdaptiveExtendScheme(a, b, 5, grid, no_initial_splitting=False, version=0, automatic_extend_split=True, split_single_dim=True, operation=operation)

# perform testcases
algorithmArray = [(adaptiveCombiInstanceExtend, 1, 2, errorOperator, 'Auto-Extend scheme (single-dim.), min=1, max=2, split=5'),(adaptiveCombiInstanceExtend2, 1, 2, errorOperator, 'Auto-Extend scheme, min=1, max=2, split=5')]
performTestcaseArbitraryDim(f, a, b, algorithmArray, 4, dim, 7, grid, 1, 6)
```
